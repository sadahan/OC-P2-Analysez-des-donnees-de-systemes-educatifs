## Pourcentage des valeurs nulles par colonne

```python
def build_nan_barplot(df_list, df_names, figsize=(17,5), color_xticklabels=False, labels_to_color=[]):
    '''
    Fonction pour tracer un ou plusieurs barplots qui représente(nt) la proportion 
    de valeurs nulles par colonnes dans le(s) dataframe(s)
    '''
    fig = plt.figure(figsize=figsize)

    for df, name in zip(df_list, df_names):
        fig.add_subplot(1, len(df_list), df_names.index(name) + 1)
        nb_nan = df.isna().sum() * 100 / len(df)
        ax = nb_nan.plot(kind='bar')
        ax.set_ylabel('Null values (%)')
#         plt.text(s=name, x=0, y=100, color='red')
        plt.title('Valeurs manquantes dans la table ' + name, fontsize=20)
        
        if color_xticklabels:
            for xtick in ax.get_xticklabels():
                if xtick.get_text() in labels_to_color[df_names.index(name)]:
                    xtick.set_color('red')

                
build_nan_barplot([EdStatsFootNote, EdStatsCountry_Series], ['EdStatsFootNote','EdStatsCountry_Series'])
build_nan_barplot([EdStatsCountry, EdStatsSeries], ['EdStatsCountry','EdStatsSeries'], color_xticklabels=True,
                 labels_to_color=[['Country Code', 'Region', 'Income Group'], ['Series Code', 'Topic', 'Indicator Name', 'Long definition']])
build_nan_barplot([EdStatsData], ['EdStatsData'])
# Que les données de 2000 et plus
build_nan_barplot([EdStatsData.drop(EdStatsData.loc[:,'1970':'1999'],1)], ['EdStatsData'])
```

- Modification de EdStatsData : division en 3 datasets (projections, DataCountry, DataRegions)

```python
## on ne garde que les données entre 2000 et 2100

# EdStatsData.drop(EdStatsData.iloc[:,4:34],1,inplace=True)
EdStatsData.drop(EdStatsData.loc[:,'1970':'1999'],1,inplace=True)

## On divise EdStatsData pour avoir un dataframe avec les données par pays et un avec les données par région

DataRegions = EdStatsData[~EdStatsData['Country Code'].isin(EdStatsCountry['Country Code'].tolist())]
DataCountry = EdStatsData[EdStatsData['Country Code'].isin(EdStatsCountry['Country Code'].tolist())]

## On assigne la région et l'income group à chaque pays

DataCountry = EdStatsCountry.merge(DataCountry, on='Country Code')
```

```python
## on renomme "high income OECD" et "high income NON OECD" parce qu'ils n'ont pas le même nom dans EdStatsData et EdStatsCountry 

DataCountry['Income Group'] = np.where(DataCountry['Income Group'].str.startswith('High income'),
                                       np.where(DataCountry['Income Group'] == 'High income: OECD', 'OECD members', 'High income'),
                                       DataCountry['Income Group'])
```

```python
## les données de 2020 et plus sont des projections que l'on met de côté

# print(set(DataCountry[~DataCountry['2020'].isnull()]['Indicator Name'].tolist()))

projections = DataCountry[~DataCountry['2020'].isnull()]

## on retire les projections de DataCountry et DataRegions

DataCountry = DataCountry[DataCountry['2020'].isnull()]
DataCountry.drop(DataCountry.loc[:,'2017':'2100'],1,inplace=True)
DataRegions = DataRegions[DataRegions['2020'].isnull()]
DataRegions.drop(DataRegions.loc[:,'2017':'2100'],1,inplace=True)
```

Fonction pour récupérer la donnée la plus récente et supprimer les colonnes vides :

```python
def get_last_value(df_to_treat):
    df = df_to_treat.copy(deep=True)
    df['last_value_year'] = df.apply(lambda x: x.last_valid_index(), axis=1)
    df['last_value'] = df.apply(lambda x: x[x.last_value_year], axis=1)
    return df.dropna(how='all', axis=1)
```

```python
## on calcule la population ayant accès à internet de 2011 à 2016

data_access = access[['Country Code', '2011','2012','2013','2014','2015','2016']].set_index('Country Code')
data_pop = population[['Country Code', '2011','2012','2013','2014','2015','2016']].set_index('Country Code')

def pop_with_internet(df, pop):
    for i, row in df.iterrows():
        for year in range(2011,2017):
            if not math.isnan(df.loc[i,str(year)]) and not math.isnan(pop.loc[i,str(year)]):
                df.loc[i,str(year)] = round(df.loc[i,str(year)] /100 * pop.loc[i,str(year)])
    return df

access_internet = pop_with_internet(data_access, data_pop).reset_index()
access_internet = access_internet.merge(access[['Country Name', 'Country Code']], on='Country Code')
access_internet
```

L'utilisation du quantile pour déterminer le score permet de minimiser l'impact des outliers contrairement à un score basé sur la moyenne par exemple. Et il permet d'avoir un score toujours compris entre 0 et 1 plutôt que plusieurs échelles.

```python
## On attribue le score d'accès à internet
from sklearn.preprocessing import QuantileTransformer

qt = QuantileTransformer(n_quantiles=len(access_internet)) 
access_internet['score_it'] = qt.fit_transform(access_internet[['2016']])
```

```python
## on calcule la croissance de ces indicateurs par région

regions_enrolment = test_regions[test_regions['Indicator Code'].isin(['SE.SEC.ENRL', 'SE.TER.ENRL'])].rename(
    columns={'Country Name': 'Region'}).set_index(['Region', 'Indicator Code']).drop(
    columns=['Country Code', 'Indicator Name', '2015', '2016'])
regions_enrolment = regions_enrolment.pct_change(axis=1)
```

```python
## on utilise la croissance par région pour remplir les valeurs manquantes

def fill_missing_values(df, growth):
    '''
    Fonction pour remplir les valeurs manquantes en prenant la moyenne entre la croissance de l'indicateur
    par région géographique et la croissance de l'indicateur par catégorie de revenus.
    '''
    for i, row in df.iterrows():
        for year in range(2001,2015):
            if math.isnan(df.loc[i,str(year)]) and not math.isnan(df.loc[i,str(year-1)]):
                avg_growth = (growth.loc[(row['Income Group'], row['Indicator Code']), str(year)] + growth.loc[(row['Region'], row['Indicator Code']), str(year)]) / 2
                df.loc[i,str(year)] = df.loc[i,str(year-1)] + df.loc[i,str(year-1)] * avg_growth
    df = round(df)
    return df
    
enrolment = fill_missing_values(test_enrolment, regions_enrolment)
enrolment = enrolment[['Country Name', '2009', '2010', '2011', '2012', '2013', '2014']].groupby('Country Name').sum().reset_index()
enrolment
```

## Score 4 : Taux d'inscrits dans le tertiaire

- Indicateur : taux de personnes à l'université par rapport au nombre de personnes qui pourraient y être (SE.TER.ENRR 'Gross enrolment ratio, tertiary, both sexes (%)' )

```python
enrolment_ratio = DataCountry_filtered[DataCountry_filtered['Indicator Code'].isin(['SE.TER.ENRR'])]
enrolment_ratio = get_last_value(enrolment_ratio)
print('Pas de données :', set([x for x in access['Country Name'].tolist() if x not in enrolment_ratio['Country Name'].tolist()]))

print(enrolment_ratio.count())
## comme pour l'indicateur précédent, les données ont beaucoup de valeurs manquantes
```

```python
## on n'utilise pas pct_change car les données sont déjà en %
## on calcule juste la différence de % entre les années

def calc_perc_diff(df, first_year, last_year):
    for i, row in df.iterrows():
        for year in range(last_year, first_year, -1):
            df.loc[i,str(year)] = (df.loc[i,str(year)] - df.loc[i,str(year-1)])
    return df

regions_enrol_ratio = test_regions[test_regions['Indicator Code'].isin(['SE.TER.ENRR'])].rename(
    columns={'Country Name': 'Region'}).set_index('Region').drop(
    columns=['Country Code', 'Indicator Name', 'Indicator Code','2015', '2016'])
regions_enrol_ratio = calc_perc_diff(regions_enrol_ratio, 2000, 2014)

regions_enrol_ratio
```

```python
def fill_missing_values_ratio(df, growth):
    for i, row in df.iterrows():
        for year in range(2001,2015):
            if math.isnan(df.loc[i,str(year)]) and not math.isnan(df.loc[i,str(year-1)]):
                avg_growth = (growth.loc[row['Income Group'], str(year)] + growth.loc[row['Region'], str(year)]) / 2
                df.loc[i,str(year)] = df.loc[i,str(year-1)] + avg_growth
    return df

enrolment_ratio = fill_missing_values_ratio(enrolment_ratio, regions_enrol_ratio)
enrolment_ratio = enrolment_ratio[['Country Name', '2009', '2010', '2011', '2012', '2013', '2014']].copy()
enrolment_ratio
```

```python
print('moyenne enrolment:', enrolment_ratio['2014'].mean(), '\nmediane:', enrolment_ratio['2014'].median(),
      '\necart-type:', enrolment_ratio['2014'].std())
print('quartiles:', s.quantiles(enrolment_ratio['2014'].tolist(), n=4))
```

```python
qt = QuantileTransformer(n_quantiles=len(enrolment_ratio)) 
enrolment_ratio['score_enrolment(%)'] = qt.fit_transform(enrolment_ratio[['2014']])
```

## Croissance des indicateurs

On calcule le TCAM (Taux de Croissance Annuel Moyen) pour chaque indicateur sur les 5 dernières années de données disponibles.

```python
def multiply_rates(x, last_year):
    '''
    Fonction pour calculer le taux de croissance global
    (produit des coefficients multiplicateurs).
    '''
    result = x[str(last_year)]
    for i in range(last_year-4,last_year):
         result = result * x[str(i)]
    return result

def get_growth_rate(df, last_year, name, ratio=False): 
    '''
    Fonction pour calculer le taux de croissance annuel moyen sur 5 ans
    lorsque les données sont en quantité ou en pourcentage (ratio=True).
    '''
    df_growth = df.copy(deep=True).set_index('Country Name').fillna(method='bfill', axis=1)
    five_years = str(int(last_year) - 5)
    if ratio:
        df_growth = calc_perc_diff(df_growth, int(five_years), int(last_year)).drop(columns=[five_years])
    else:       
        df_growth = df_growth.loc[:,five_years:last_year]
        df_growth = df_growth.pct_change(axis=1).drop(columns=[five_years])
    df_growth = df_growth.apply(lambda x : x/100 + 1) ## coeff multiplicateur
    df_growth['global_rate'] = df_growth.apply(lambda x : multiply_rates(x, int(last_year)), axis=1) ## taux de croissance global
    df_growth['TCAM_' + name] = df_growth.apply(lambda x: (x['global_rate'] ** (1/5) -1)*100 , axis=1)
    df_growth.reset_index(inplace=True)
    return df_growth
```

```python
pib_growth = get_growth_rate(pib, '2016', 'pib')
access_growth = get_growth_rate(access, '2016', 'internet')
enrol_growth = get_growth_rate(enrolment, '2014', 'nb_students').fillna(0)
enrol_ratio_growth = get_growth_rate(enrolment_ratio, '2014', 'enrolment', ratio=True)
test_growth = pib_growth[['Country Name', 'TCAM_pib']].merge(
            access_growth[['Country Name', 'TCAM_internet']], on='Country Name').merge(
            enrol_growth[['Country Name', 'TCAM_nb_students']], on='Country Name').merge(
            enrol_ratio_growth[['Country Name', 'TCAM_enrolment']], on='Country Name')
test_growth[test_growth['Country Name'].isin(liste_final)]
```

```python
def build_growth_lineplot(df, ylabel, title, liste_final):
    '''
    Fonction pour tracer le graphe de l'évolution d'un indicateur.
    '''
    
    df = df[df['Country Name'].isin(liste_final)].set_index('Country Name')
    sns.set(rc={'figure.figsize':(10,10)})
    sns.lineplot(data=df.T)
    plt.legend(bbox_to_anchor=(1,1), loc="upper left")
    plt.ylabel(ylabel, fontsize = 15)
    plt.title(title, y=1, fontsize=20);
```
